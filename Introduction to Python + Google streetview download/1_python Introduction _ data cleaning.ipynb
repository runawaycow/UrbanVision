```python
# use pd.read_csv()
df = pd.read_csv('F:\\_QU-TIanjin studio\\gis\\point.csv',encoding='utf-8') 
print(len(df))
df.head(3)
```

```python
# select the columns
df_final = df[['ID','lon',"lat",'angle','highway']]

df_final.head(2)
              
```

```python
## we can also:
df_final.replace('motorway_link', 'motorway',inplace=True)
df_final.replace('primary_link', 'primary', inplace=True)
df_final.replace('secondary_link', 'secondary', inplace=True)
df_final.replace('tertiary_link', 'tertiary', inplace=True)
df_final.replace('trunk_link', 'trunk', inplace=True)
```

```python
## if you have buffer id:
# df_final2 = df_final.drop_duplicates(['buffer_ID'],keep = 'first').copy()
# df_final2 = df_final2.reset_index(drop=True)

df_final.columns = ['ID','Lng',"Lat",'angle','highway']
```

# Visualization

```python
## 2.1 import the library
import folium
import io
from PIL import Image

```

```python
#2.2 Get the center point of our sample locations as the center of the map
Center = [df_final['Lat'].mean(),df_final['Lng'].mean()]
print(Center)
```

```python
# 2.3 creating a basemap that’s centered to our sample, defining styles, and zoom in scale

map0 = folium.Map(location=Center,tiles='Stamen Toner',zoom_start=12)
map0.save("SVI_Map.html")
# Display the map0
map0
```

```python
# 2.4 Create Color band
import branca.colormap as cm

print(df_final.angle.max())
print(df_final.angle.min())
```

```python

clr=['green', 'yellow', 'red']
colormap = cm.LinearColormap(clr,vmin=0, vmax=360) # Change the value domain here
colormap
```

```python
# 2.5 Paint the dataframe data with dots on the map
# To save time, we subsample 500 random pts from the 1500 as a new datafram

df1=df_final.sample(1500)

# Read the Lat and Lng and paint with color
for index, row in df1.iterrows():
    folium.CircleMarker([row['Lat'], row['Lng']],radius=0.3,fill=True,
                        color=colormap(row['angle']),opacity=0.8,
                        fill_color=colormap(row['angle']),fill_opacity=0.8,
                        ).add_to(map0)  

    colormap.add_to(map0)

# Display the map
map0
```
